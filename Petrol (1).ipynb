```python
#importing datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data set

df=pd.read_csv('C:/Users/kenyowilliams/Downloads/Petrol Dataset June 23 2022 -- Version 2.csv', encoding='latin1')
df
```

**https://www.kaggle.com/datasets/zusmani/petrolgas-prices-worldwide?select=Petrol+Dataset+June+23+2022+--+Version+2.csv**

**Petrol/Gas prices worldwide.
It contains world share of petrol consumption by country, price per gallon, price per litter and the corresponding price in Pakistan Rupee (Conversion Rate 1 USD = 211.5 PKR)**

```python
#Data cleaning
#Dropping Column
df=df.drop('S#',axis=1)
#Reseting index
df=df.reset_index(drop=True)
#NA's
df.isna().sum()
#Column Rename
df=df.rename(columns= {'World Share':'World Share (%)'})
#Symbols
df['Daily Oil Consumption (Barrels)']=df['Daily Oil Consumption (Barrels)'].str.replace(',','')
df['World Share (%)']=df['World Share (%)'].str.replace('%','')
df['GDP Per Capita ( USD )']=df['GDP Per Capita ( USD )'].str.replace(',','')
df['Gallons GDP Per Capita Can Buy']=df['Gallons GDP Per Capita Can Buy'].str.replace(',','')
#Rounding
df['Yearly Gallons Per Capita']=round(df['Yearly Gallons Per Capita'],0).astype(int)
df
```

```python
#Daily Oil Consumption (Gallons)
df['Daily Oil Consumption (Gallons)']=round(df['Daily Oil Consumption (Barrels)'].astype(int)/(55),0)
df
```

```python
#Sort By Price
df.sort_values(by='Price Per Gallon (USD)',ascending=False,inplace=False).head(5)
```

**Initially I tried to sort on Daily Oil Consumption in terms of barrels and it yeilded suspicious results. The likes of the United States, China, and India were not included which tipped me that something was going on here. I decided to create a new column that captures Daily Oil Consumption in terms of Gallons and sort on this value so that this analysis is still viable.**

```python
#Sorting by newly calculated column
df.sort_values(by='Daily Oil Consumption (Gallons)',ascending=False,inplace=False).head(5)
```

```python
#Graphing Top 5 Countries in terms of Price Per Gallon
df2=df.sort_values('Price Per Gallon (USD)',ascending=False,inplace=False)
priceplot=plt.figure()
priceplot.set_figwidth(7)
priceplot.set_figheight(7)
ax=priceplot.add_axes([0,0,1,1])
ax.bar(df2['Country'].head(10),df2['Price Per Gallon (USD)'].head(10),color='red')
ax.set_yticks((0,5,10,15,20,25,30,35,40,45,50,55))
plt.title('Top 10 Price Per Gallon (USD)',fontsize=30,color='blue')
plt.ylabel('Price Per Gallon (USD)',fontsize=15)
ax.margins(x=0.01)
plt.show()
```

**This graph makes sense fr the most part. Hong Kong and Norway making top five is a little more shocking. The other nations on this list are not because of their small roles in global trading activities. However, after doing a little more research on the unit of measurement used in both Hong Kong and Norway, they both use litres as liquid measurement standard. I searched Wikipedia to see if there was a distinct difference between these countries in terms of vehicle usage. Shockingly Norway has only 200 less Motor Vehicles per 1000 than the United States does. After digging a little deeper into an explanation for Norway I was able to uncover a bit of a breakdown in what components go into setting their gas price points. It seems as though alot of their gas price is made up of high taxes that are collected by the government to fund publically financed things in their country. Another explanation why this graph may not be telling the whole story, is the difference in currency between the United States and Norway. 1 Norweigian Krone roughly equals 0.95 United States Dollars. This means that the puchasing power that citizens of Norway enjoy is higher than the United States so these prices might not be all that high in the eyes of Norweigian citizens.**

```python
#Creating a graph for Consumption of Oil (Gallons) by Country
df3=df.sort_values(by='Daily Oil Consumption (Gallons)',ascending=False,inplace=False)
consumeplot=plt.figure()
consumeplot.set_figwidth(5)
consumeplot.set_figheight(5)
ax=consumeplot.add_axes([0,0,1,1])
ax.bar(df3['Country'].head(10),df3['Daily Oil Consumption (Gallons)'].head(10),color='blue')
plt.title('Top 10 Daily Oil Consumption (Gallons) Countries',color='red',fontsize=25)
plt.ylabel('Daily Oil Consumption (Gallons)',fontsize=15)
plt.xticks(fontsize=7)
ax.margins(x=0.01)
plt.show()
```

**This graph isnt very shocking to me at all. All of these countries are relatively developed nations with large GDP values and a large amount of population. Doubling back on the vehicle ownership totals by country on Wikipedia, China has the most total motor vehicles in the world, followed by the United States and India. I feel as though the Oil consumption is higher in the United States than it is in China due to the popularity of trucks and larger sized Vehicles than in China. Since China has such a large population, I think there is less space in many of their cities to build infrastructure large enough to fit both pedestrians and large sized vehicles. Another asset that I think that plays into the UNited states having the largest daily demand for oil is that many Americans do not utilize the public transportation available to them for their own respective reasons. This is almost opposite of the situations that you find in the other top 4 countries. I found Japans position on this chart interesting as well. Japan is the smallest country in the top five in terms of land area as well as population. What I suspect is fueling their surge to top five is their highly developed industrial sector. For 2021 Japan had roughly 40,000 USD in GDP per capita. That is almost half of the United States's GDP Per capita and four to eight times larger than every other country in this top five. Heavy reliance on oil machinery equipment that yeilds more effective production processes is the only reasonable means by which Japan can output at this rate, thus explaining their high usage of oil on a daily basis.**
