```python
#packages+popout plots
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
#loading dataset 

df=pd.read_csv('C:/Users/kenyowilliams/Downloads/covid_worldwide.csv')
df.head()
```

```python
#Cleaning data (dropping NA) & removing commas 
df['Total Cases']=df['Total Cases'].str.replace(',','')
df['Total Deaths']=df['Total Deaths'].str.replace(',','')
df['Total Recovered']=df['Total Recovered'].str.replace(',','')
df['Active Cases']=df['Active Cases'].str.replace(',','')
df['Total Test']=df['Total Test'].str.replace(',','')
df['Population']=df['Population'].str.replace(',','')
df=df.replace(np.NaN,0)
df
```

```python
#Addition of Calculated Rows to Add context
df['% of Population covered by Tests']=round((df['Total Test'].astype(int)/df['Population'].astype(int)*100),2).replace(np.nan,0).replace(np.inf,0)
df['% of Deaths by Population']=round((df['Total Deaths'].astype(int)/df['Population'].astype(int)*100),2).replace(np.nan,0).replace(np.inf,0)
df['% of Cases by Population']=round((df['Total Cases'].astype(int)/df['Population'].astype(int)*100),2).replace(np.nan,0).replace(np.inf,0)
df
```

```python
#Using Sorting Methods to further uncover trends by country 
#Its important to note here that some countries (China) in specific had alot of there reporting information filled in as NaN values
#Thus this led to my calculated columns returning infinty, to fix this i filled inf values with zero so focus remained on available data
#Tests First 
df=df.sort_values(by=['% of Population covered by Tests'],ascending=False)
df.head(10)
```

**This sort tells us is that tests were widely available. As you can see above, these countries are vastly different in terms of location, population, overall infection and assumed wealth yet all of them have almost if not more than 1000% of their population covered by the testing kits that they had available**

```python
#Now Deaths
df=df.sort_values(by=['% of Deaths by Population'],ascending=False)
df.head(10)
```

**This sort displays a rather interesting trend. It seems to be the case that the variant strains that were at large in these countries were more lethal than in other areas of the world. A potenital contributor to the high death rates in these countries could be the medical infrastructure in place in these areas. It can be observed that comparatively speaking to the last sort that we executed with testing, these areas have a rather low percentage of their population covered with available tests. It has been widely prescribed that early detection and isolation is one of the best ways to combat death from Covid.**

```python
#Lastly Cases
df=df.sort_values(by=['% of Cases by Population'],ascending=False)
df.head(10)
```

**This sort helps to affirm my earlier suspicion that poor medical infrastructure plays a big role in the spread of Covid in countries. Minus Gibraltar and Austria, these countries had a low percentage of their population covered by availabile  testing. Geographically speaking this makes sense though. When looking at a map of Gibraltar and Slovenia's population density, you can see that a large amount of their population is within a small region. As you move futher and further away from this region, you see this population denisty taper off rather drastically. So it could be the case that a large amount of the testing was available in urban areas and not in rural areas. Those who were infected in rural areas may have been infected at a higher clip than those who lived in urban areas if the testing materials were not evenly distributed throughout the country**
