## Passing Data Frame

### **Scrape + Constructing Data Frame**

```python
Pass1Scrape=pd.read_html('https://www.nfl.com/stats/player-stats/')
Pass2Scrape=pd.read_html('https://www.nfl.com/stats/player-stats/category/passing/2023/REG/all/passingyards/DESC?aftercursor=AAAAGQAAABlAeBAAAAAAADFleUp6WldGeVkyaEJablJsY2lJNld5SXpPRFVpTENJek1qQXdORGMwTVMwMU1qTTJMVE15TWpVdE9UQmxOaTAxWW1aa1ltRTBNVEEyTkdNaUxDSXlNREl6SWwxOQ==')
Pass1=Pass1Scrape[0]
Pass2=Pass2Scrape[0]
Pass=pd.concat([Pass1,Pass2],axis=0,sort=False,copy=False,ignore_index=False)
Pass.reset_index(inplace=True)
Pass=Pass.drop({'index','Sck','SckY'},axis=1)
display(HTML(Pass.to_html(index=False)))
```

### Calculating and adding Fantasy Score to Data Frame

```python
Pass['Fantasy Score']=((4*Pass['TD'])+((Pass['Pass Yds'])/25)-(4*Pass['INT']))
Pass=Pass.sort_values(by=['Fantasy Score','TD','Pass Yds'],ascending=False)
display(HTML(Pass.to_html(index=False)))
```

## **Kickers**

**Scraping+Constructing Data Frame**

```python
KickScrape1=pd.read_html('https://www.nfl.com/stats/player-stats/category/field-goals/2023/reg/all/kickingfgmade/desc')
KickScrape2=pd.read_html('https://www.nfl.com/stats/player-stats/category/field-goals/2023/REG/all/kickingfgmade/DESC?aftercursor=AAAAGQAAABZACAAAAAAAADFleUp6WldGeVkyaEJablJsY2lJNld5SXpJaXdpTXpJd01EUmtOVGt0TkRVME1DMDNNalV6TFRrME9XWXROalZsTkRjME1qSTJZekF4SWl3aU1qQXlNeUpkZlE9PQ==')
Kick1=KickScrape1[0]
Kick2=KickScrape2[0]
nums={'0','1','3','4'}
Kicking=pd.concat([Kick1,Kick2],axis=0,sort=False,copy=False,ignore_index=False)
display(HTML(Kicking.to_html(index=False)))
```

### **Cleaning up the Data Frame to get rid of fractions**

```python
Kicking['1-19 > A-M']=Kicking['1-19 > A-M'].str.replace('/',' ',regex=True)
Kicking['1-19 > A-M']=Kicking['1-19 > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['1-19 > A-M']=Kicking['1-19 > A-M'].astype(int)

Kicking['20-29 > A-M']=Kicking['20-29 > A-M'].str.replace('/',' ',regex=True)
Kicking['20-29 > A-M']=Kicking['20-29 > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['20-29 > A-M']=Kicking['20-29 > A-M'].astype(int)

Kicking['30-39 > A-M']=Kicking['30-39 > A-M'].str.replace('/',' ',regex=True)
Kicking['30-39 > A-M']=Kicking['30-39 > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['30-39 > A-M']=Kicking['30-39 > A-M'].astype(int)

Kicking['40-49 > A-M']=Kicking['40-49 > A-M'].str.replace('/',' ',regex=True)
Kicking['40-49 > A-M']=Kicking['40-49 > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['40-49 > A-M']=Kicking['40-49 > A-M'].astype(int)

Kicking['50-59 > A-M']=Kicking['50-59 > A-M'].str.replace('/',' ',regex=True)
Kicking['50-59 > A-M']=Kicking['50-59 > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['50-59 > A-M']=Kicking['50-59 > A-M'].astype(int)

Kicking['60+ > A-M']=Kicking['60+ > A-M'].str.replace('/',' ',regex=True)
Kicking['60+ > A-M']=Kicking['60+ > A-M'].map(lambda x: x.lstrip('0123456789101112131415161718192021222324252627282930313233343536373839404142434445'))
Kicking['60+ > A-M']=Kicking['60+ > A-M'].astype(int)
```

### **Calculating and adding missed kicks to Data Frame**

```python
Kicking['Missed Kicks']=Kicking['Att']-Kicking['FGM']
```

### **Calculating and adding Fantasy Points to Data Frame**

```python
Kicking['Fantasy Score']=((6*Kicking['60+ > A-M'])+(5*Kicking['50-59 > A-M'])+(4*Kicking['40-49 > A-M'])
                        +(3*(Kicking['30-39 > A-M']+Kicking['20-29 > A-M']+Kicking['1-19 > A-M'])))-Kicking['Missed Kicks']
Kicking=Kicking.sort_values(by=['Fantasy Score'],ascending=False)
display(HTML(Kicking.to_html(index=False)))
```
