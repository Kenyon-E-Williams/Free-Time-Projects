**Setting the initial scraper up**

```python
scraper=pd.read_html('https://en.wikipedia.org/wiki/List_of_best-selling_video_games')
```

**Figuring out which of the tables on this URL that I want to focus in on**

```python
for idx,table in enumerate(scraper):
    print('*********************************************************************************************')
    print(idx)
    print(table)
```

**Now that I have established that I want to focus on the table in the 1st index position I will call it**

```python
VideoGameList=scraper[1]
VideoGameList
```

```python
NACount=VideoGameList.isna().sum()
Overview=VideoGameList.info()
SalesInfo=VideoGameList['Sales'].astype(float).apply(lambda x: '{:.4f}'.format(x)).describe()
#SalesInfo=VideoGameList.describe(include = [np.number])
print(f"{NACount}\n{Overview}\n{SalesInfo}")
```
